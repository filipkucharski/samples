```python
import requests

url = "https://cloud.iexapis.com/stable/stock/aapl/chart/3m?token=pk_6ff506dbc200413b919921b4e8023a45"
api_data = requests.get(url).json()
```

```python
apple_stock_df = pd.DataFrame.from_dict(api_data)
```

```python
apple_stock_df["date"] = pd.to_datetime(apple_stock_df["date"] , format = "%Y-%m-%d")
```
